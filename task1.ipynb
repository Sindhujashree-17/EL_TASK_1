```python
#Read dataset
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/netflix_titles[1].csv")
df
```

```python
df.describe()
```

```python
#Checking for missing values
df.isnull()
```

```python
print("Missing values before cleaning:\n")
df.isnull().sum()
```

```python
#Checking for duplicates
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
#Standardize text values
df["type"] = df["type"].str.strip().str.title()
df["country"] = df["country"].str.strip().str.title()
df["rating"] = df["rating"].str.strip().str.upper()
df
```

```python
#Convert date formats
df["date_added"] = pd.to_datetime(df["date_added"], errors = "coerce")
df["date_added"] = df["date_added"].dt.strftime("%d-%m-%Y")
df
```

```python
# Renaming column headers to lowercase
df.columns =  df.columns.str.lower().str.strip().str.replace(" ", "_")
df
```

```python
# Fixing data types
df["release_year"] = df["release_year"].astype(int)
df["title"] = df["title"].astype(str)
df["type"] = df["type"].astype(str)
df
```

```python
# Handling missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('NR', inplace=True)  # NR = Not Rated
df['duration'].fillna('Unknown', inplace=True)
df['date_added'].fillna('01-01-2000', inplace=True)
```

```python
print("\nMissing values after cleaning:\n", df.isnull().sum())
```

```python
# Checking for data types
print("\nData types after cleaning:\n", df.dtypes)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each country and get the top 10
country_counts = df['country'].value_counts().nlargest(10)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
country_counts.plot(kind='barh', color='skyblue')
plt.title('Top 10 Countries Producing Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.gca().invert_yaxis()  # Show highest values at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
